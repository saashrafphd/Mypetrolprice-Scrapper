```python
# =============================================================================
# Import library
# =============================================================================

browser = webdriver.Chrome(options=option,service=ChromeService(ChromeDriverManager().install()))


html = "https://www.mypetrolprice.com/2/Diesel-price-in-Delhi?FuelType=1&LocationId=2"
city_name = 'Delhi'
Data = []
# =============================================================================
# Get HistroyData
# =============================================================================
"""Get Diesel Price"""
browser.get(html)
pages = int(
    browser.find_element('xpath',
        '//*[@id="BC_GV_CustomGridPager_LabelNumberOfPages"]'
    ).text
)
print(pages)
" Get Table in a perticular Page"
for p in range(1, pages+1):
    print(f"Page:{p} city_name: {city_name}")
    # //*[@id="BC_GV_CustomGridPager_NextButton"]
    NEXT_BUTTON_XPATH = '//input[@type="submit" and @name="ctl00$BC$GV$ctl07$CustomGridPager$NextButton" and @value=">"]'
    try:
        elem = browser.find_element('xpath',NEXT_BUTTON_XPATH)
        table = browser.find_element(By.ID,"BC_GV")
        # Entire Row
        rows = table.find_elements(By.TAG_NAME ,"td")
        for i in range(0, (len(rows) - 1)):
            # Append Price
            Price = rows[i].find_element(By.CLASS_NAME,"GVPrice").text.replace("₹ ", "")
            # Append Date
            Date = re.sub("\n", "-", rows[i].find_element(By.CLASS_NAME,"DateDiv").text)
            Date = pd.to_datetime(Date).strftime("%Y-%m-%d")
            # Append city_name
            Data.append([city_name,Date,float(Price)])
            time.sleep(3)
        #print(Data)
        elem.click()        
    except:
        print("Exit may be last page")
        browser.close()
    # browser.forward()
browser.close()

```

```python
df = pd.DataFrame(Data, columns = ['City', 'Date','Price'])  
print(df.shape)
df.to_csv(f'C://Users//Syed Abdullah Ashraf//Desktop//delhi diesel price.csv', index= False)
df
```

```python
diesel_urls = []
with open('page_source.html', 'w', encoding="utf-8") as f:
        f.write(browser.page_source)
        with open('page_source.html','r',encoding="utf-8") as o:
            html_soup = BeautifulSoup(o,'html.parser')
            data_containers = html_soup.find_all('div',class_='sixteen columns row')
            for containers in data_containers:
                states = containers.find_all('h2')
                #print(states)
                cities= containers.find_all('div',class_='CH')
                #print(cities)
                for city in cities:
                    #print(city)
                    for a in city.find_all('a',{'class':''}):
                        diesel_urls.append([a.text,a['href']])
                        
```

```python
diesel_url_df = pd.DataFrame(diesel_urls, columns=['city','link'])
diesel_url_df
```

```python
def proper_url(x):
    z = re.findall(r'\d+', x)[0]
    return x+str('?FuelType=0&LocationId=')+str(z)

diesel_url_df['url'] = diesel_url_df['link'].apply(lambda x : proper_url(x))
diesel_url_df.head()
```
